```python
# subseting the original Dataframe
df_subset = df[["school", "sex", "address", "Pstatus","guardian", "studytime", "internet", "famrel", "freetime", "goout", "absences", "G1", "G2", "G3", "romantic"]]
df
df_subset.head()
```

We choose categorical variables and turn them into dummy variables. The way we do so is by creating a list and loop around it to change the names of the new column and place them in a single Dataframe.

```python
list_to_dummies = df_subset[["school", "sex", "address", "Pstatus", "guardian", "studytime", "internet", 
                            "famrel", "goout", "romantic"]].columns.tolist()

list_to_dummies
```

```python
# looping through the list and naming the variables for proper distinction
df_temp = pd.DataFrame()
for item in list_to_dummies:
    dummy = pd.get_dummies(df_subset[item])
    dummy_cols = dummy.columns.tolist()
    for cols in dummy_cols:
        dummy = dummy.rename(columns={cols : str(item)+"_"+str(cols)})
    df_temp = pd.concat([df_temp,dummy], axis = 1)
df_temp.head()
```

```python
#adding quantitative variables to our new dataframe 
df_temp = pd.concat([df_temp, df_subset.loc[:,"G1":"G3"]], axis = 1)
df_temp = pd.concat([df_temp, df_subset["romantic"]], axis = 1)
df_temp
```

```python
# checking the value count of romantic variable
df_temp["romantic"].value_counts()
```

```python
# changing the variable from yes and no to 0 and 1 for later use
def change_yesno_to_numbers(x):
    if x == "yes":
        return("1")
    else:
        return("0")

df_temp["romantic"] = df_temp["romantic"].apply(change_yesno_to_numbers)
df_temp["romantic"] = df_temp["romantic"].astype(int)
df_temp["romantic"]
```

```python
df_temp.info()
```

In order to avoid colinearity and dummy variable trap, we select _n - 1_ where _n_ is the number of dummy variable for every category

```python
# instead of selecting each variable, we put all the column names in a list and removing the ones we dont need
list_of_X = df_temp.columns.tolist()
list_of_X.remove("school_MS")
list_of_X.remove("sex_M")
list_of_X.remove("address_U")
list_of_X.remove("Pstatus_T")
list_of_X.remove("guardian_other")
list_of_X.remove("studytime_4")
list_of_X.remove("internet_yes")
list_of_X.remove("famrel_5")
list_of_X.remove("goout_5")
list_of_X.remove("romantic_yes")
list_of_X.remove("romantic_no")
list_of_X.remove("romantic")
X = df_temp[list_of_X]
y = df_temp["romantic"]
```

# Knn Model

for this model we are going to use K Nearest Neighbor algorithm

Knn algorithm identifies new observations based on their  Eucledian distance from their nearest neighbors. In our case we are going to choose _k = 3_

```python
# spliting the model into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# creating an instance of the knn model from sklearn library
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
# training the model
knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

our Knn score is **0.625**, which means **62%** of our model prediction is accurate

```python
# creating a confusion matrix 
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test) 
cm = confusion_matrix(y_test, y_pred) 
cm = pd.DataFrame(cm, index=['No (actual)', 'Yes (actual)'], columns = ['No (pred)', 'Yes (pred)']) #label and make df
cm
```
